## Line Plot

### Univariate
Visual representation of the relationship of a single column in a dataset.

```python
# Group by release year and count titles
yearly_count = netflix_df['release_year'].value_counts().sort_index()

# Plotting
plt.plot(yearly_count.index, yearly_count.values)
plt.title("Number of Movie/TV Show Released per Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Movie/TV Show")
plt.show()
```

```python
# adding gird-lines
plt.grid(True)

# Plotting
plt.plot(yearly_count.index, yearly_count.values)
plt.title("Number of Movie/TV Show Released per Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Movie/TV Show")
# Set x-axis for every 10 units
plt.yticks(range(0, max(yearly_count.index) + 1, 100))  
plt.show()
```

```python
#Bar Plot
#Count by type
type_count = netflix_df['type'].value_counts()

# Plotting
plt.bar(yearly_count.index, yearly_count.values)
plt.title("Number of Movies vs. TV Shows")
plt.xlabel("Release Year")
plt.ylabel("Number of Movie/TV Show")
plt.show()
```

## Scatter Plot

### Bivariate
Visual representation of the relationship between two variables in a dataset.

```python
# Convert duration to numeric (handling errors)
netflix_df['duration_numeric'] = pd.to_numeric(netflix_df['duration'].str.extract('(\d+)')[0], errors='coerce')

# Plotting
plt.scatter(netflix_df['release_year'], netflix_df['duration_numeric'])
plt.title("Release Year vs. Duration")
plt.xlabel("Release Year")
plt.ylabel("Duration (Minutes or Seasons)")
plt.show()
```

### Multivariate
Visual representation of the relationship between three or more columns/ variables in a dataset.

```python

netflix_df['duration_numeric'] = pd.to_numeric(netflix_df['duration'].str.extract('(\d+)')[0], errors='coerce')

# Mapping content types to colors
content_type_color = {"Movie": "red", "TV Show": "blue"}

for content_type, color in content_type_color.items():
    # Filter data by content type
    content_data = netflix_df[netflix_df["type"] == content_type]
    
    # Scatter plot for each content type
    plt.scatter(content_data["release_year"], content_data["duration_numeric"], c=color, label=content_type)

plt.grid(True)
plt.title("Netflix: Release Year vs. Duration (Movies vs. TV Shows)")
plt.xlabel("Release Year")
plt.ylabel("Duration (Minutes or Seasons)")
plt.legend()
plt.show()
```

## Box Plots

```python
# Plotting box plot to compare durations by type
sns.boxplot(x='type', y='release_year', data=netflix_df,palette="Set3")
plt.title("Box Plot: Duration by Content Type")
plt.xlabel("Type")
plt.ylabel("Duration (Minutes or Seasons)")
plt.show()
```

## Kernel Density Plot (KDE)

KDE plot helps to visualize the distribution of data in a dataset



```python
# Convert duration to numeric (handling errors)
netflix_df['duration_numeric'] = pd.to_numeric(netflix_df['duration'].str.extract('(\d+)')[0], errors='coerce')

# Set the figure size
plt.figure(figsize=(10, 6))
sns.kdeplot(data=netflix_df[netflix_df['type'] == 'Movie']['duration_numeric'], 
            label='Movies', 
            fill=True, 
            color='blue', 
            alpha=0.6)

sns.kdeplot(data=netflix_df[netflix_df['type'] == 'TV Show']['duration_numeric'], 
            label='TV Shows', 
            fill=True, 
            color='green', 
            alpha=0.6)

# Adding titles and labels
plt.title("KDE Plot: Duration Distribution for Movies and TV Shows")
plt.xlabel("Duration (Minutes or Seasons)")
plt.ylabel("Density")
plt.legend()
plt.show()
```

## Violin Plot (Box Plot + KDE)

A statistical graphic that shows the distribution of numerical data by combining a box plot and a kernel density plot.

```python
# Plotting violin plot to visualize the distribution of duration by type
sns.violinplot(x='type', y='release_year', data=netflix_df)
plt.title("Violin Plot: Distribution of Release year by Content Type")
plt.xlabel("Type")
plt.ylabel("Release Year")
plt.show()
```

## Correlation Heat Maps

```python
# Calculate correlation matrix for numerical columns
correlation_matrix = netflix_df[['release_year', 'duration_numeric']].corr()

# Plotting the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap: Release Year and Duration")
plt.show()
```
